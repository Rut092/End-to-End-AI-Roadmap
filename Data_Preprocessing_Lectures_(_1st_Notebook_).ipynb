**Taking Care of Missing Data**
* generally we have alot of data if
1% data is missing is good to remove it.
* sometimes we have alot of missing data we have to do it right way
we will do missing value of all the average of all the column

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(x[:,1:])
# make sure to input only numericals values not string fields it does not process string field
imputer.transform(x[:,1:])
x[:,1:] = imputer.transform(x[:,1:])
print(x)
```

1. The Real-World Analogy: The Exam
Think of the imputer like a student preparing for an exam.

>* fit (The Study Phase): The student looks at the textbook (the data) and calculates the average (mean). They now "know" that the average height is 170cm. They haven't changed anything yet; they just have that number stored in their head.

>* transform (The Exam Phase): The student sees a blank spot on the exam paper. Because they "know" the average is 170cm from their study phase, they fill in the blank with "170".

> If you didn't fit first, the student wouldn't know what number to use to fill the gap.

2. The reason is consistency between Training and Testing data.

>* On Training Data: You use fit to calculate the mean. Then you use transform to fill the holes.

>* On New/Test Data: You must not fit again. You want to use the same mean you calculated from the training data to fill holes in the new data.

> If you fit on the test data, you are "cheating" by letting the model see the patterns in the test set before it's supposed to.

**Key Rule: You fit only once (on training data), but you transform everything (training data, test data, and future real-world data).**


```python
# Encoding Categorical Data
# turn countries(spain,germany,France) into number 0,1,2,....etc
# In this dataset we have 3 countries if we turn them into sequence (0,1,2)
# it will think there is order between these countries and it will think that order matters
# but that is not the case that's why we want to avoid misinterpreted correlation between features and outcome
# we can do ONEHOTENCODING --> turning (3 countries in a Column to 3 Different Columns of binary vectors )
```

```python

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# The first column (index 0) is 'Country', which is categorical
# We want to apply OneHotEncoder to this column
# remainder='passthrough' means that all other columns not specified will be kept as they are
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')

x = np.array(ct.fit_transform(x))

print(x)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

**Splitting the dataset into Training set and Test set**

```python
# we should apply feature scaling after the splitting not before !!
# we want all feature to apply into a same scale we do this to overcome
# one feature dominating the other one !
```

```python
# if we do before the test dataset will get all the mean and standard
# deviation of the whole dataset we dont want that to happen
# it will cause information leakage
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```
